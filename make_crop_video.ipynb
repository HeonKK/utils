```python

def make_video_from_frame(video_name, frame_path, fps, fourcc, width, height):
    """
    frame to video using cv2.VideoWriter
    
    video_path : output video save path
    
    fourcc : frame code
    fps : frame per second
    width, height : video size that you want to make
    """
    fourcc = cv2.VideoWriter_fourcc(*'DIVX')
    out = cv2.VideoWriter(filename=video_name, fourcc=fourcc, fps = fps,
                          frameSize = (width, height))

    files = [f for f in os.listdir(frame_path) if isfile(join(frame_path, f))]
    files.sort(key = lambda x: x[5:-4])
    files.sort()
    for i in range(len(files)):    
        frame = cv2.imread(frame_path + files[i], cv2.IMREAD_COLOR)
        out.write(frame)   
    
    out.release()

def video_to_frame(video_path, frame_path):
    """
    video file to image file
    """
    vidcap = cv2.VideoCapture(video_path)

    success = True
    count = 1

    while success:
        success, image = vidcap.read()


        if image is None:
            break;

        cv2.imwrite(frame_path+"%05d.jpg" % count, image)
        print("\rSave %d Frames to " % count + frame_path, end="")
        count += 1
    print("\nVideo to Frame Done!")
    
def video_to_frame_30_to_25(video_path, frame_path, duration):
    
    vidcap = cv2.VideoCapture(video_path)

    fps = round(vidcap.get(cv2.CAP_PROP_FPS), 1)
    
    length = (duration) * 25


    count = 0
    skip_count = 1

    success = True

    while success:
        success, image = vidcap.read()
    
        if image is None or count == length:
            break;
        if skip_count % 6 == 0:
            skip_count += 1
            continue
        
        cv2.imwrite(frame_path+"%05d.jpg" % count, image)
        print("\rSave %d Frames to " % count + frame_path, end="")
        
        count+=1
        skip_count+=1   
```
